```python
import re
import pandas as pd
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://thepointsguy.com/guide/best-surfing-destinations-europe/#:~:text=The%209%20best%20surfing%20destinations%20in%20Europe%201,has%20a%20far-flung%20feel%20and%20some%20windy%20'

response = requests.get(url)

response


response.content

soup=BeautifulSoup(response.content)

soup

title = soup.find_all('h2')
```

```python
surf_list= []
for i in title[:9]:
    surf_list.append(i.text)
    


surf_loc = []
for dirt in surf_list:
    z = re.sub('\d+\.\s','',dirt)
    surf_loc.append(z)
print(surf_loc)


countries = []
places = []
for i in surf_loc:
    z=i.split(', ')
    places.append(z[0])
    countries.append(z[1])



df = pd.DataFrame({'City':places,'Country':countries})
df
```

```python
url2 = 'https://www.bestsurfdestinations.com/best-places-to-surf-in-europe/'

response = requests.get(url2)

response


response.content

soup=BeautifulSoup(response.content)

soup

title2 = soup.find_all('h2')

surf_loc2 = []
for i in title2:
    surf_loc2.append(i.text)


print(surf_loc2)

countries2 = []
places2 = []
for i in surf_loc2:
    z=i.split(', ')
    places2.append(z[0])
    countries2.append(z[1])


df2 = pd.DataFrame({'City':places2,'Country':countries2})
df2

```

```python
import re
url3 = 'https://www.travelinglifestyle.net/amazing-surfing-spots-in-europe-when-to-go/'

response = requests.get(url3)

response


response.content

soup=BeautifulSoup(response.content)

soup
rule = re.compile('^h-')
title3 = soup.find_all('h2',attrs={'id': rule})

surf_loc3 = []
for i in title3:
    surf_loc3.append(i.text)

print(surf_loc3)

countries3 = []
places3 = []
for i in surf_loc3:
    z=i.split(', ')
    places3.append(z[0].capitalize())
    countries3.append(z[1])

df3 = pd.DataFrame({'City':places3,'Country':countries3})
df3
```

```python
url = 'https://www.lonelyplanet.com/articles/epic-surf-spots-europe'

response = requests.get(url)

response


response.content

soup=BeautifulSoup(response.content)

soup

title4 = soup.find_all('h2')

surf_loc4 = []
for i in title4[:9]:
    surf_loc4.append(i.text)

countries4 = []
places4 = []
for i in surf_loc4:
    z=i.split(', ')
    places4.append(z[0])
    countries4.append(z[1])

df4 = pd.DataFrame({'City':places4,'Country':countries4})
df4
```

```python
final_df=pd.concat([df,df2,df3,df4 ], ignore_index=True)

final_df2 =final_df.drop_duplicates()

final_df2.to_csv('final_df.csv')

```
